```python
import json
```

```python
PROMPT_DICT = {
    "prompt_input": (
        "Below is an instruction that describes a task, paired with an input that provides further context. "
        "Write a response that appropriately completes the request.\n\n"
        "### Instruction:\n{instruction}\n\n### Input:\n{input}\n\n### Response:"
    ),
    "prompt_no_input": (
        "Below is an instruction that describes a task. "
        "Write a response that appropriately completes the request.\n\n"
        "### Instruction:\n{instruction}\n\n### Response:"
    ),
}
```

```python
PROMPT_DICT_Simple = {
    "prompt_input": (
        #"Below is an instruction that describes a task, paired with an input that provides further context. "
        #"Write a response that appropriately completes the request.\n\n"
        "Instruction:\n{instruction}\n\nInput:\n{input}\n\n"
    ),
    "prompt_no_input": (
        #"Below is an instruction that describes a task. "
        #"Write a response that appropriately completes the request.\n\n"
        "Instruction:\n{instruction}\n\n"
    ),
}
```

```python
with open("alpaca_data.json", "r") as f:
    data = json.load(f)

```

```python
all_prompts=[]
for item in data:
    ann=item
    if ann.get("input", "") == "":
        prompt = PROMPT_DICT_Simple["prompt_no_input"].format_map(ann)
    else:
        prompt = PROMPT_DICT_Simple["prompt_input"].format_map(ann)
    all_prompts.append(prompt)
```

```python
all_prompts
```

```python
all_dialogs=[]
for item in all_prompts:
    dict_item={"role": "user", "content": item}
    all_dialogs.append([dict_item])
    
output_file_path = "all_dialogs.json"

# 使用 json.dump() 将数据保存为 JSON 文件
with open(output_file_path, "w") as output_file:
    json.dump(all_dialogs, output_file)

print(f"Data saved to {output_file_path}")
```
